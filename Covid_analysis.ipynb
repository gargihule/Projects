```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('/content/covid19.xlsx')
```

```python
df.info()
```

## **UNIVARIATE ANALYSIS**

### **1.Total COVID Cases in India**

```python
df['Total Cases'].sum()
```

**2. Total Recovered Patients**

```python
df['Discharged'].sum()
```

**3.Histogram Graph**

```python
plt.figure(figsize=(8,5))

sns.histplot(df['Total Cases'], kde=True)

plt.title('Distribution of Total Cases')

plt.xlabel('Total Cases')
plt.ylabel('Frequency')

plt.show()
```

**4.Active Cases Analysis**

```python
df['Active'].value_counts()
```

**5.Deaths Analysis**

```python
df['Deaths'].value_counts()
```

### **6.Total Cases Analysis**

```python
df['Total Cases'].value_counts()
```

## **7.Pie Chart**

```python
top5 = df.nlargest(5, 'Total Cases')

plt.figure(figsize=(7,7))

plt.pie(top5['Total Cases'],
        labels=top5['State/UTs'],
        autopct='%1.1f%%')

plt.title("Top 5 States by Total Cases")

plt.show()
```

## **8.Distplot**

```python
plt.figure(figsize=(8,5))

sns.distplot(df['Total Cases'])

plt.title("Distribution Plot of Total Cases")

plt.xlabel("Total Cases")

plt.show()
```

# **BIVARIATE ANALYSIS**

### **1. Total Cases State-wise [Barplot]**

```python
state_cases = df.groupby('State/UTs')['Total Cases'].sum()

state_cases.plot(kind='bar',
                 figsize=(12,6))

plt.xlabel("States/UTs")
plt.ylabel("Total Cases")
plt.title("State-wise Total COVID Cases")

plt.xticks(rotation=90)

plt.show()
```

**2. States Having Cases Greater Than 1,00,000**

```python
df[df['Total Cases'] > 100000]
```

**3. Top 5 States with Highest Deaths**

```python
df.nlargest(5, 'Deaths')[['State/UTs', 'Deaths']]
```

**4.Top 5 States with Highest Cases**

```python
df.nlargest(5, 'Total Cases')[['State/UTs', 'Total Cases']]
```

**5. Scatter Plot**

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Total Cases',
                y='Deaths',
                data=df)

plt.title('Total Cases vs Deaths')

plt.show()
```

# **MULTIVARIATE ANALYSIS**

**1. State-wise Total Cases and Deaths**

```python
df.groupby(['State/UTs', 'Deaths'])['Total Cases'].sum()
```

**2. State-wise Active and Discharged Cases**

```python
df.groupby(['State/UTs', 'Active', 'Discharged'])['Total Cases'].sum()
```

**3. Heatmap Graph**

```python
plt.figure(figsize=(10,6))
sns.heatmap(df[['Total Cases',
                'Active',
                'Discharged',
                'Deaths',
                'Population']].corr(),
                annot=True,
            cmap='coolwarm')
plt.title('Multivariate Correlation Heatmap')
plt.show()
```

**4. Population and Death Analysis**

```python
df.groupby(['Population', 'Deaths'])['Total Cases'].sum()
```

**lineplot**

```python
plt.figure(figsize=(10,5))

sns.lineplot(x='Total Cases',
             y='Discharged',
             data=df)

plt.title("Total Cases vs Discharged")

plt.show()
```

pairplot

```python
sns.pairplot(df[['Total Cases',
                 'Active',
                 'Discharged',
                 'Deaths']])

plt.show()
```
